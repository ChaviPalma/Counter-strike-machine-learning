## Ranking de modelos

1. Support Vector Machine:

Este modelo resultó ser el mejor de todos, ya que obtuvo los errores más bajos tanto en el Mean Absolute Error como en el Root Mean Squared Error. En otras palabras, hace predicciones muy precisas. Además, su R² de 0.830 indica que puede explicar alrededor del 83% de la variabilidad en los datos, lo que habla muy bien de su capacidad para ajustarse a lo que realmente pasa con los MatchKills, sin mucha dispersión. También hay que decir que escalar los datos antes de entrenarlo ayudó bastante, ya que así pudo identificar mejor las relaciones no lineales. En resumen, un modelo bastante sólido y eficiente.

2. Random Forest:

Es un modelo bastante bueno porque consigue que tanto el Mean Absolute Error como el Root Mean Squared Error sean bajos, lo que significa que tiene una buena precisión al hacer predicciones. Su R² (coeficiente de determinación) es de 0.813, lo que indica que logra explicar el 81.3% de la variabilidad en los datos. Esto sugiere que sus predicciones están bastante alineadas con la realidad de los MatchKills. Pero también se deja ver una ligera tendencia al sobreajuste, algo a tener en cuenta

3. Decision Tree:

Es un modelo aceptable, aunque no tan preciso como los dos anteriores. Sus valores de Mean Absolute Error y Root Mean Squared Error son moderadamente bajos, lo que indica que tiene un nivel de precisión decente, pero algo por debajo de los mejores modelos. Su R² (coeficiente de determinación) es de 0.763, lo que significa que explica el 76.3% de la variabilidad en los datos. En general, logra captar la mayoría de las relaciones entre las variables de entrada y los MatchKills, aunque también muestra cierta tendencia al sobreajuste, por lo que no generaliza tan bien como los otros modelos

4. Regresión lineal múltiple:

Es un modelo decente, aunque con menor precisión en comparación con los modelos no lineales. Sus valores de Mean Absolute Error y Root Mean Squared Error son un poco más altos, lo que indica que sus predicciones no se ajustan tan bien a los valores reales. Su R² (coeficiente de determinación) es de 0.710, lo que significa que logra explicar alrededor del 71% de la variabilidad en los datos. Esto muestra que el modelo es capaz de captar relaciones lineales entre las variables de entrada y los MatchKills, pero tiene dificultades para identificar relaciones más complejas que podrían mejorar su rendimiento general

5. Regresión Lineal Simple:

Este es el modelo con el rendimiento más bajo entre todos los que evaluamos. Sus valores de Mean Absolute Error y Root Mean Squared Error son los más altos, lo que significa que sus predicciones se desvían bastante de los valores reales. Además, su R² (coeficiente de determinación) de 0.550 indica que solo logra explicar alrededor del 55% de la variabilidad en los datos. Esto deja en evidencia que el modelo es bastante limitado, ya que utiliza solo una variable predictora, lo cual le impide captar las relaciones más complejas que existen entre las variables de entrada y los MatchKills.

## ¿Por qué Support vector machine es el mejor modelo?

*   Alta precisión: Tiene una alta precisión, ya que obtuvo los valores más bajos de Mean Absolute Error (MAE) y Root Mean Squared Error (RMSE). Esto significa que sus predicciones estuvieron muy cerca de los valores reales, lo cual demuestra que el modelo tiene una capacidad excepcional para predecir con bastante exactitud la cantidad de MatchKills.
*  Alto R² (Coeficiente de determinación): Con un R² de 0.830, el modelo SVM logra explicar el 83% de la variabilidad en los datos. Esto demuestra que tiene una muy buena capacidad para capturar la relación entre las variables predictoras y la variable objetivo de forma sólida y confiable.
*   Captura de relaciones no lineales: Gracias al uso de kernels como el RBF o el polinómico, el modelo SVM puede detectar y modelar relaciones no lineales y complejas entre las variables. Esto le da una clara ventaja frente a modelos más simples, como la regresión lineal, que solo pueden captar relaciones lineales
*   Eficacia con escalado de datos: Al escalar los datos antes de entrenar el modelo, el SVM puede manejar mejor las diferencias en las magnitudes de las variables predictoras, lo que ayuda a optimizar su rendimiento y a obtener resultados más precisos
